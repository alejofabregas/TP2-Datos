```python
from pyspark import SparkContext
from pyspark.sql import SQLContext
from pyspark.sql import SparkSession
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
sqlContext = SQLContext(sc)

crew = sqlContext.read.parquet('/content/drive/MyDrive/Organización de Datos/TP2/Movie dataset/crew.parquet').rdd
```

```python
crew_mas_trabajo = crew.map(lambda x: (x.crew_id, (x.department, x.id))).distinct().map(lambda x: (x[0], 1)).reduceByKey(lambda a,b: a+b).takeOrdered(1, lambda x: -x[1])
crew_mas_trabajo = crew_mas_trabajo[0]
#Con el distinct se filtran aprox 8000 tuplas, de 464314 a 456764

crew_mas_trabajo_nombre = crew.filter(lambda x: x.crew_id == crew_mas_trabajo[0]).first().name

print("El crew que trabajó en más tuplas (departamento, película) distintas es", crew_mas_trabajo_nombre, "con", crew_mas_trabajo[1], "tuplas")
```
