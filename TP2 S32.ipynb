```python
from pyspark import SparkContext
from pyspark.sql import SQLContext
from pyspark.sql import SparkSession
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
sqlContext = SQLContext(sc)

crew = sqlContext.read.parquet('/content/drive/MyDrive/Organización de Datos/TP2/Movie dataset/crew.parquet').rdd
movies = sqlContext.read.parquet('/content/drive/MyDrive/Organización de Datos/TP2/Movie dataset/movies.parquet').rdd
```

```python
crew_por_pelicula = crew.map(lambda x: (x.id, 1)).reduceByKey(lambda a,b: a+b).sortBy(lambda x: x[1]).cache()

cant_pelis = crew_por_pelicula.count()

mediana = crew_por_pelicula.zipWithIndex().filter(lambda x: x[1] == cant_pelis // 2).first()

print("La mediana de cantidad de crew para las películas es", mediana[0][1])
```
