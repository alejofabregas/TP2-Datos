```python
from pyspark import SparkContext
from pyspark.sql import SQLContext
from pyspark.sql import SparkSession
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
spark = SparkSession.builder.getOrCreate()
sc = spark.sparkContext
sqlContext = SQLContext(sc)

actores = sqlContext.read.parquet('/content/drive/MyDrive/Organización de Datos/TP2/Movie dataset/imdb_actors.parquet').rdd
```

```python
frecuencias_años_muerte = actores.map(lambda x: (x.deathYear, 1)).filter(lambda x: x[0] != '\\N').reduceByKey(lambda a,b: a+b)

#año_mas_mortal = frecuencias_años_muerte.max(lambda x: x[1])
año_mas_mortal = frecuencias_años_muerte.reduce(lambda a,b: a if a[1] > b[1] else b)

print("El año más mortal para actores y actrices fue el", año_mas_mortal[0], "con", año_mas_mortal[1], "muertes")
```
